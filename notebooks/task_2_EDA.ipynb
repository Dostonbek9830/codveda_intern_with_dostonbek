```python
import pandas as pd 
```

```python
df = pd.read_csv('../data/iris.csv')
df.head()
```

```python
# Check for missing values
df.isna().sum()
df.info()
```

```python
#change data types
df['species'] = df['species'].astype('category')
```

```python
# Show duplicated rows
df[df.duplicated(keep=False)]
```

```python
# Remove duplicates
df.drop_duplicates(subset=None, keep='first', inplace=True)
```

```python
# Ensure duplicates are removed
df[df.duplicated(keep=False)]
```

```python
# Clean 'species' column by stripping whitespace and converting to lowercase
df['species'] = df['species'].str.strip().str.lower()
```

```python
#Starting exploratory data analysis
df.describe()
```

```python
#Visualize the distribution of each feature using histograms
import matplotlib.pyplot as plt
df['sepal_length'].hist()
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
df['petal_length'].hist(color='green')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
df['petal_width'].hist(color='lightblue')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
df.hist(column=['sepal_width', 'petal_length', 'petal_width', 'sepal_length'], figsize=(10, 8), color='orange', edgecolor='black')
plt.suptitle('Distribution of Sepal Width, Petal Length, and Petal Width')
plt.show()
```
